## Dataset taken from UCI repository

```python
# sample data is given
# here we will practice about list comprehension style for reading data from dataframe

# importing libaries
import pandas as pd
import numpy as np
```

```python
# data is in csv format. So if any data is not properly available, while reading the data we will avoid it. 
# Therefore, we put error_bad_lines=False
data = pd.read_csv('data.csv',',',error_bad_lines=False)
```

```python
data.head(6)
```

```python
# check null values in total
data.isnull().sum()
```

```python
# check the null value position
data[data['password'].isnull()]
```

```python
# drop null value
data.dropna(inplace=True)
```

```python
# convert the data into array
passwords_tuple=np.array(data)
```

```python
passwords_tuple
```

```python
 #shuffling randomly for robustness
import random
random.shuffle(passwords_tuple)
```

```python
# list comprehension
# considering first index as label
y=[labels[1] for labels in passwords_tuple]
```

```python
y
```

```python
X=[labels[0] for labels in passwords_tuple]
```

```python
# plot
import seaborn as sns
sns.set_style('whitegrid')
sns.countplot(x='strength',data=data,palette='RdBu_r')
```

```python
data.shape
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
print(word_divide_char('foood'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
# logistic regression classifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
```

```python
# ovr = one vs all
log_class=LogisticRegression(penalty='l2',multi_class='ovr')
log_class.fit(X_train,y_train)
```

```python
print(log_class.score(X_test,y_test))
```

```python
## Multinomial
clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train) #training
print(clf.score(X_test, y_test))
```

```python
# check
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```
